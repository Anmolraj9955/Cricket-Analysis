**batting_summary**

```python
with open(r"E:\CRICKET ANALYSIS PROJECT\t20_json_files\t20_wc_batting_summary.json") as f:
    data = json.load(f)

all_records = []
for rec in data:
    all_records.extend(rec['battingSummary'])

df_batting = pd.DataFrame(all_records)
df_batting.head(20)
    
```

```python
df_batting.tail()
```

```python
df_batting["out/not_out"] = df_batting.dismissal.apply(lambda x: "out" if len(x)>0 else "not_out") 
```

```python
df_batting.head()
```

```python
df_batting.head(10)
```

```python
df_batting.drop(columns = ["dismissal"], inplace = True)
df_batting.head(10)
```

```python
df_batting['batsmanNames'] = df_batting['batsmanName'].apply(lambda x: x.replace('â€', '' ))
df_batting['batsmanNames'] = df_batting['batsmanName'].apply(lambda x: x.replace('\xa0', '' ))
df_batting.head(11)
```

```python
df_batting.drop(columns = ['batsmanNames'], inplace = True)
```

```python
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('â€', '' ))
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('\xa0', '' ))
df_batting.head(11)
```

***lets add these two tables using fk***

```python
match_ids_dict = {}

for index, row in df_match.iterrows():
         key1 = row['team1'] + ' Vs ' + row['team2']
         key2 = row['team2'] + ' Vs ' + row['team1']

         match_ids_dict[key1] = row["match_id"]
         match_ids_dict[key2] = row["match_id"]
match_ids_dict   
```

```python
df_batting['match_id'] = df_batting['match'].map(match_ids_dict)
df_batting.head()
```

```python
df_batting.to_csv(r'E:\CRICKET ANALYSIS PROJECT\t20_csv/fact_batting_summary.csv', index = False)
```
